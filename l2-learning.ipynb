````python
data = pd.read_csv('data.csv');
X = data.get(data.columns[0]);
Y = data.get(data.columns[1]);
N = float(len(X))
````

````python
L = 0.0001#learning rate
h = 1.5#hyperparameter value
epoch = 100
````

````python
w = 0;b = 0;
errors = [];

# here now we train the model. hurrah!
def l2_learning():
  global w,b
  for i in range(epoch):
    Y_hat = w*X + b
    errors.append((2/N) * sum(Y - Y_hat)**2 - h * w**2)
    D_m = (-4/N) * sum(X * (Y - Y_hat)) - 2 * h * w
    D_c = (-4/N) * sum(Y - Y_hat) - h
    w = w - L * D_m
    b = b - L * D_c
````

Here we did L2 Regression (y = wx + b). The loss function is :

```
2/N * sum(y - (wx + b))**2 - h * sum(w**2)

where h = hyperparameter,
      N = total number of records
```

in order to minimize the loss function, we need to find values if w and b that satisfy the equation best. This can be done by calculating d/dw and d/db on the data


```
d/dw = -4/N * sum(x * (y - (wx + b))) - 2 * h * w

d/db = -4/N * sum(y - (wx + b)) - h
```





````python
l2_learning()
Y_predicted = w*X + b
print('w is ', w, ' and b is', b)


fig, (plot1, plot2) = plt.subplots(ncols = 2,gridspec_kw={'right':2})

plot1.grid()
plot1.scatter(X, Y)
plot1.plot(X, Y_predicted, color='red')
plot1.set_title('Scatter Plot')
plot1.set_xlabel('Area in Meters Sq.')
plot1.set_ylabel('Price in \'000 Ksh');

plot2.grid()
plot2.plot(list(range(epoch)), errors)
plot2.set_title('Error Plot')
plot2.set_xlabel('Iteration')
plot2.set_ylabel('Error');
````
